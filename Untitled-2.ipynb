```python
import matplotlib.pyplot as plt
import numpy as np
import math

def load_file(filename):
    f = open(filename, 'r')
    coords = []
    for line in f:
        if line.startswith('ATOM'):
            x = float(line[30:38])
            y = float(line[38:46])
            z = float(line[46:54])
            t = str(line[13:17])
            if t == 'N   ' or t == 'CA  ' or t == 'C   ':
                coords.append(np.array([x, y, z]))
    return coords
    
def angle(v1, v2):
    n1 = np.sqrt(sum(v1 * v1))
    n2 = np.sqrt(sum(v2 * v2))
    s = sum(v1 * v2) / (n1 * n2)
    return np.arccos(max(-1, min(s, 1)))

def cross_product(v1, v2):
    a, b, c = v1
    d, e, f = v2
    c1 = np.linalg.det(np.array(((b, c), (e, f))))
    c2 = -np.linalg.det(np.array(((a, c), (d, f))))
    c3 = np.linalg.det(np.array(((a, b), (d, e))))
    return np.array([c1, c2, c3])

# From https://www.math.fsu.edu/~quine/MB_10/6_torsion.pdf
def calc_dihedral(v1, v2, v3, v4):
    ab = v1 - v2
    cb = v3 - v2
    db = v4 - v3
    u = cross_product(ab, cb)
    v = cross_product(db, cb)
    w = cross_product(u, v)
    a = angle(u, v)
    if angle(cb, w) > 0.001:
        a = -a
    return a

def get_phi_psi(coords):
    phi = []
    psi = []
    for i in range(4, len(coords) - 3, 3):
        #(c, n, ca, c)
        phi.append(calc_dihedral(coords[i-2], coords[i-1], coords[i], coords[i+1]))
        #(n, ca, c, n)
        psi.append(calc_dihedral(coords[i-1], coords[i], coords[i+1], coords[i+2]))
    return zip(phi, psi)

def draw_plot(filename):
    x = []
    y = []
    coords = load_file(filename)
    phi_psi = get_phi_psi(coords)
    for phi, psi in phi_psi:
        x.append(math.degrees(phi))
        y.append(math.degrees(psi))
    
    plt.xlim([-180, 180])
    plt.ylim([-180, 180])
    plt.scatter(x, y, s = 5)
    plt.xlabel('phi')
    plt.ylabel('psi')
    plt.grid()
    plt.show()
```

```python
draw_plot('6lu7.pdb')
```

```python
draw_plot('7c2e.pdb')
```
